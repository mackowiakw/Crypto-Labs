```python
%matplotlib inline

import cv2
import random
from matplotlib import pyplot as plt
```

```python
img_source = cv2.imread('pudzian.jpg', cv2.IMREAD_GRAYSCALE)

for i in range(len(img_source)):
  for j in range(len(img_source[0])):
    img_source[i][j] = 0 if img_source[i][j] < 128 else 255

plt.imshow(img_source, cmap='gray')
plt.show()
```

```python
img_source
```

```python
img_1 = [ [0] * len(img_source[0]) for _ in range(len(img_source))]
img_2 = [ [0] * len(img_source[0]) for _ in range(len(img_source))]

for i in range(len(img_source)):
  for j in range(len(img_source[0])):

    if img_source[i][j] == 0:
      if random.random() < 0.5:
        img_1[i][j] = 0
        img_2[i][j] = 0
      else:
        img_1[i][j] = 255
        img_2[i][j] = 255

    else:
      if random.random() < 0.5:
        img_1[i][j] = 0
        img_2[i][j] = 255
      else:
        img_1[i][j] = 255
        img_2[i][j] = 0


img_res = [ [0] * len(img_source[0]) for _ in range(len(img_source))]

for i in range(len(img_source)):
  for j in range(len(img_source[0])):
    if img_1[i][j] == img_2[i][j]:
      img_res[i][j] = 0
    else:
      img_res[i][j] = 255
```

```python
fig = plt.figure(figsize=(25,10))

imgs = [img_source, img_1, img_2, img_res]
titles = ['Original', 'Part A', 'Part B', 'Combined']

for id, (img, title) in enumerate(zip(imgs, titles)):
  ax = fig.add_subplot(1, 4, id+1)
  ax.set_title(title, y=-0.2, pad=-14)
  plt.imshow(img, cmap='gray')


plt.show()
```
